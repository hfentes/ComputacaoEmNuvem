```python
#declarando diretório principal
ROOT_DATA_PATH = "drive/MyDrive/data/"
```

```python
#lendo arquivo movies.csv informando o separador
m_df = pd.read_csv(f'{ROOT_DATA_PATH}/movies.csv', sep=';')
#limpando linhas com valor invalido se existir
m_df.dropna()
m_df.head()
```

```python
#setando os nomes das colnas
nomes_colunas = ['Movie_Id', 'Title_Year']
m_df.columns = nomes_colunas
m_df
```

```python
#lendo o arqivo customers_rating.csv informando q tem cabecalho e o separador
cr_df = pd.read_csv(f'{ROOT_DATA_PATH}/customers_rating.csv',header=0, sep=';')
#limpando linhas com valor invalido se existir
cr_df.dropna()
cr_df.info()
```

```python
# 1.1. Quantos filmes estão disponíveis no dataset?

#pegando qtds dos IDs dos filmes de forma unica
qtd_filmes = m_df['Movie_Id'].nunique()
print(f"O número de filmes disponíveis no dataset é {qtd_filmes}.")
```

```python
#1.2. Qual é o nome dos 5 filmes com melhor média de avaliação?

# Mesclando os dataframes
merged_df = pd.merge(cr_df, m_df, on='Movie_Id')

# Calculando a média de avaliação para cada filme
mean_ratings = merged_df.groupby(['Movie_Id'])['Rating'].mean()

# Ordenando os filmes com base na média de avaliações
sorted_ratings = mean_ratings.sort_values(ascending=False)

# Obtendo os 5 primeiros filmes com a melhor média de avaliação
top_5_movies = pd.merge(sorted_ratings, m_df, on='Movie_Id').head(5)

print("Os 5 filmes com a melhor média de avaliação são:")
print(top_5_movies)
```

```python
#1.3. Quais os 9 anos com menos lançamentos de filmes?

# Extraindo o ano de lançamento a partir da coluna 'título e ano de lançamento'
m_df['Year'] = m_df['Title_Year'].str[-5:-1]

# Contando o número de lançamentos por ano
count_per_year = m_df['Year'].value_counts()

# Ordenando os anos com base no número de lançamentos
sorted_years = count_per_year.sort_values()

# Obtendo os 9 anos com menos lançamentos de filmes
bottom_9_years = sorted_years.head(9)

print("Os 9 anos com menos lançamentos de filmes são:")
print(bottom_9_years)
```

```python
#1.4. Quantos filmes que possuem avaliação maior ou igual a 4.7, considerando apenas os filmes avaliados na última data de avaliação do dataset?

# Convertendo a coluna 'Date' para o formato de data
cr_df['Data'] = pd.to_datetime(cr_df['Date'])

# Encontrando a última data de avaliação
ultima_data = cr_df['Data'].max()

# Filtrando os filmes avaliados na última data de avaliação
filmes_ultima_data = cr_df[cr_df['Data'] == ultima_data]

# Calculando a média das avaliações para cada filme da selecao
media_avaliacoes_por_filme = filmes_ultima_data.groupby('Movie_Id')['Rating'].mean()

# Filtrando os filmes com uma média de avaliação maior ou igual a 4.7
filmes_com_alta_media = media_avaliacoes_por_filme[media_avaliacoes_por_filme >= 4.7]

# Contando o número de filmes com média de avaliação maior ou igual a 4.7
num_filmes_com_alta_media = len(filmes_com_alta_media)

print(f"O número de filmes que possuem média de avaliação maior ou igual a 4.7 no dataset é {num_filmes_com_alta_media}.")

```

```python
#1.5. Dos filmes encontrados na questão anterior, quais são os 10 filmes com as piores notas e quais as notas?

# Obtendo os filmes da questão anterior
# Mesclando os dataframes para trazer os nomes dos filmes
filmes_com_nomes = pd.merge(filmes_com_alta_media, m_df, on='Movie_Id')

# Ordenando os filmes com base no rating em ordem ascendente
piores_filmes = filmes_com_nomes.sort_values('Rating').head(10)

# Selecionando apenas as colunas necessárias
piores_filmes = piores_filmes[['Title_Year', 'Rating']]

print("Os 10 filmes com as piores notas e seus nomes são:")
print(piores_filmes)
```

```python
#1.6. Quais os id's dos 5 customers que mais avaliaram filmes e quantas avaliações cada um fez?

# Obtendo os 5 clientes que mais avaliaram filmes
top_5_customers = cr_df['Cust_Id'].value_counts().head(5)

print("Os 5 clientes que mais avaliaram filmes e suas respectivas contagens de avaliações são:")
print(top_5_customers)
```
