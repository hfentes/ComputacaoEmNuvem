```python
#declarando diret√≥rio principal
ROOT_DATA_PATH = "drive/MyDrive/data/"
```

```python
#lendo arquivo movies.csv informando o separador
m_df = pd.read_csv(f'{ROOT_DATA_PATH}/movies.csv', sep=';')
#limpando linhas com valor invalido se existir
m_df.dropna()
m_df.head()
```

```python
#setando os nomes das colnas
nomes_colunas = ['Movie_Id', 'Title_Year']
m_df.columns = nomes_colunas
m_df
```

```python
#lendo o arqivo customers_rating.csv informando q tem cabecalho e o separador
cr_df = pd.read_csv(f'{ROOT_DATA_PATH}/customers_rating.csv',header=0, sep=';')
#limpando linhas com valor invalido se existir
cr_df.dropna()
cr_df.info()
```

```python
# 1.1. Quantos filmes est√£o dispon√≠veis no dataset?

#pegando qtds dos IDs dos filmes de forma unica
qtd_filmes = m_df['Movie_Id'].nunique()
print(f"O n√∫mero de filmes dispon√≠veis no dataset √© {qtd_filmes}.")
```

```python
#1.2. Qual √© o nome dos 5 filmes com melhor m√©dia de avalia√ß√£o?

# Mesclando os dataframes
merged_df = pd.merge(cr_df, m_df, on='Movie_Id')

# Calculando a m√©dia de avalia√ß√£o para cada filme
mean_ratings = merged_df.groupby(['Movie_Id'])['Rating'].mean()

# Ordenando os filmes com base na m√©dia de avalia√ß√µes
sorted_ratings = mean_ratings.sort_values(ascending=False)

# Obtendo os 5 primeiros filmes com a melhor m√©dia de avalia√ß√£o
top_5_movies = pd.merge(sorted_ratings, m_df, on='Movie_Id').head(5)

print("Os 5 filmes com a melhor m√©dia de avalia√ß√£o s√£o:")
print(top_5_movies)
```

```python
#1.3. Quais os 9 anos com menos lan√ßamentos de filmes?

# Extraindo o ano de lan√ßamento a partir da coluna 't√≠tulo e ano de lan√ßamento'
m_df['Year'] = m_df['Title_Year'].str[-5:-1]

# Contando o n√∫mero de lan√ßamentos por ano
count_per_year = m_df['Year'].value_counts()

# Ordenando os anos com base no n√∫mero de lan√ßamentos
sorted_years = count_per_year.sort_values()

# Obtendo os 9 anos com menos lan√ßamentos de filmes
bottom_9_years = sorted_years.head(9)

print("Os 9 anos com menos lan√ßamentos de filmes s√£o:")
print(bottom_9_years)
```

```python
#1.4. Quantos filmes que possuem avalia√ß√£o maior ou igual a 4.7, considerando apenas os filmes avaliados na √∫ltima data de avalia√ß√£o do dataset?

# Convertendo a coluna 'Date' para o formato de data
cr_df['Data'] = pd.to_datetime(cr_df['Date'])

# Encontrando a √∫ltima data de avalia√ß√£o
ultima_data = cr_df['Data'].max()

# Filtrando os filmes avaliados na √∫ltima data de avalia√ß√£o
filmes_ultima_data = cr_df[cr_df['Data'] == ultima_data]

# Calculando a m√©dia das avalia√ß√µes para cada filme da selecao
media_avaliacoes_por_filme = filmes_ultima_data.groupby('Movie_Id')['Rating'].mean()

# Filtrando os filmes com uma m√©dia de avalia√ß√£o maior ou igual a 4.7
filmes_com_alta_media = media_avaliacoes_por_filme[media_avaliacoes_por_filme >= 4.7]

# Contando o n√∫mero de filmes com m√©dia de avalia√ß√£o maior ou igual a 4.7
num_filmes_com_alta_media = len(filmes_com_alta_media)

print(f"O n√∫mero de filmes que possuem m√©dia de avalia√ß√£o maior ou igual a 4.7 no dataset √© {num_filmes_com_alta_media}.")

```

```python
#1.5. Dos filmes encontrados na quest√£o anterior, quais s√£o os 10 filmes com as piores notas e quais as notas?

# Obtendo os filmes da quest√£o anterior
# Mesclando os dataframes para trazer os nomes dos filmes
filmes_com_nomes = pd.merge(filmes_com_alta_media, m_df, on='Movie_Id')

# Ordenando os filmes com base no rating em ordem ascendente
piores_filmes = filmes_com_nomes.sort_values('Rating').head(10)

# Selecionando apenas as colunas necess√°rias
piores_filmes = piores_filmes[['Title_Year', 'Rating']]

print("Os 10 filmes com as piores notas e seus nomes s√£o:")
print(piores_filmes)
```

```python
#1.6. Quais os id's dos 5 customers que mais avaliaram filmes e quantas avalia√ß√µes cada um fez?

# Obtendo os 5 clientes que mais avaliaram filmes
top_5_customers = cr_df['Cust_Id'].value_counts().head(5)

print("Os 5 clientes que mais avaliaram filmes e suas respectivas contagens de avalia√ß√µes s√£o:")
print(top_5_customers)
```
